# CUSTOM TENSORFLOW ESTIMATOR FOR SOLVING ALGEBRAIC EQUATIONS

```python
from T1 import MyDNN
import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
train1 = "train.csv"
train2 = "train2.csv"
test1 = "test.csv"
test2 = "test2.csv"
```

```python
file = pd.read_csv("algebra.csv", header=None)
```

```python
x1_normal = np.max(file[0]) - np.min(file[0])
x2_normal = np.max(file[1]) - np.min(file[1])
y1_normal = np.max(file[2]) - np.min(file[2])
```

```python
def my_input_fn(file_path, perform_shuffle = False, repeat_count = 1):
    def decode_csv(line):
        parsed_line = tf.decode_csv(line, [[0.], [0.], [0.], [0.]])
        label = parsed_line[3] # Last element is the label
        del parsed_line[-1] # Delete last element
        del parsed_line[0]
        feature_names = ["x1", "x2"]
        features = parsed_line # Everything (but last element) are the features
        features[0] = features[0] / x1_normal
        features[1] = features[1] / x2_normal
        label = label  / y1_normal
        d = [dict(zip(feature_names, features)), label]
        return d

    dataset = (tf.data.TextLineDataset(file_path) # Read text file
        #.skip(1) # Skip header row
        .map(decode_csv)) # Transform each elem by applying decode_csv fn

    if perform_shuffle:
        # Randomizes input using a window of 256 elements (read into memory)
        dataset = dataset.shuffle(buffer_size = 256)

    dataset = dataset.repeat(repeat_count) # Repeats dataset this # times
    dataset = dataset.batch(8)  # Batch size to use
    iterator = dataset.make_one_shot_iterator()
    batch_features, batch_labels = iterator.get_next()
    return batch_features, batch_labels
```

```python
fc = [tf.feature_column.numeric_column("x1"), tf.feature_column.numeric_column("x2")]
```

 ### MODEL 1 - with two hidden layer units

```python
#train the model with 85% split for the column 3 as target 
FILE = train2
my_model1 = MyDNN(feature_columns = fc, hidden_units = [2], savepoint = "./model1/")
my_model1.train(input_fn = lambda: my_input_fn(FILE, True, 8), steps = 15000)
```

```python
#train the model with 85% split for the column 4 as target 
FILE = train1
my_model2 = MyDNN(feature_columns = fc, hidden_units = [2], savepoint = "./model2/")
my_model2.train(input_fn = lambda: my_input_fn(FILE, True, 16), steps = 15000)
```

```python
#Predict the model for the train inout values to validate
print(my_model1.predict(3.40657919319416, 1.34274033191108))
```

### TEST FUNCTION - Model with 2 hidden layer units

```python
#test the model with 15% split for the column 3 as target 
FILE = test2
my_model1.evaluate(input_fn = lambda: my_input_fn(FILE, True, 8))
```

```python
#test the model with 15% split for the column 4 as target 
FILE = test1
my_model2.evaluate(input_fn = lambda: my_input_fn(FILE, True, 8))
```
